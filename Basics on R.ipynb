I used the built-in piechart function to display the pie chart with the property area, and to calculate the percentage for each component, we categorized the data and calculated the count, which we subsequently divided by the total rows in the data set.

```r
# Task 3-b Create a pie chart for Property_area and display percentages in legend respectively

data1 = as.data.frame(table(variabledt$Property_Area, dnn = list("Area")), responseName = "freq")
data1

piepercent<- round(100*data1$freq/sum(data1$freq), 1)

pie3D(data1$freq, labels = data1$Area , explode = 0.09, main = "Pie Chart")
legend("topright", legend = piepercent, cex = 1,
       fill = rainbow(length(table(variabledt$Property_Area))))
```
